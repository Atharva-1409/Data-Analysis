```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
data1 = df.loc[:, ["free", "credit", "Prediction"]]
```

```python
print("Null Values:")
print(data1.isnull().sum())
```

```python
data1 = data1.dropna()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12,6))

sns.scatterplot(
    data=data1,
    x="free",
    y="credit",
    hue="Prediction"
)

plt.title("Spam Detection using 'money' and 'business'")
plt.xlabel("Frequency of 'free'")
plt.ylabel("Frequency of 'credit'")
plt.legend(title="Prediction")

plt.show()
```
