```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Salaries.csv")
```

```python
data1 = df.loc[:, ["TotalPay"]]
```

```python
print("Null Values:")
print(data1.isnull().sum())
```

```python
data1 = data1.dropna()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(y=data1["TotalPay"])

plt.title("Distribution of Employee Total Pay")
plt.ylabel("Total Pay")

plt.show()
```
