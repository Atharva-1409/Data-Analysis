```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('test.csv')
```

```python
data1 = df.loc[:,["id","battery_power"]]
```

```python
data1.dropna()
```

```python
data1.head()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12,6))

sample = data1.sort_values(by='battery_power', ascending=False).head(10)
sns.barplot(data=sample, x='id', y='battery_power', palette='viridis')

plt.title('Battery Power by Phone ID')
plt.xlabel('Phone ID')
plt.ylabel('Battery Power')
plt.show()
```
